```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os
parent = os.path.dirname(os.getcwd())
```

```python

couverture_habitant_chemin = os.path.join(parent, 'data\output\couverture_habitant.csv')
couverture_habitant_chemin_final = couverture_habitant_chemin.replace("\\", "/")

# Chargement 
df_hab = pd.read_csv(couverture_habitant_chemin_final, dtype={'code_dep': str})

# en-tête
print("En-tête du DataFrame :")
print(df_hab.head())

# Infos structurelles
print(f"Dimensions : {df_hab.shape}")
print(df_hab.info())

# Descriptif statistique
print("\nStatistiques descriptives :")
print(df_hab.describe())

# Analyse statistique de la densité
print("\nStatistiques sur la densité (sites pour 10k hab) :")
print(df_hab['sites_pour_10k_hab'].describe())

# Qualité des données
print("\nValeurs manquantes :")
print(df_hab.isnull().sum())
print(f"Doublons détectés : {df_hab.duplicated().sum()}")

# Aperçu des valeurs extrêmes
print("\nTop 5 des départements les plus denses :")
display(df_hab.sort_values('sites_pour_10k_hab', ascending=False).head(5))
```

```python
indice_concurrence_hhi_chemin = os.path.join(parent, 'data\output\indice_concurrence_hhi.csv')
indice_concurrence_hhi_chemin_final = indice_concurrence_hhi_chemin.replace("\\", "/")

# Chargement
df_hhi = pd.read_csv(indice_concurrence_hhi_chemin_final, dtype={'code_dep': str})

# en-tête
print("En-tête du DataFrame :")
print(df_hab.head())

# Infos structurelles
print(f"Dimensions : {df_hhi.shape}")
print(df_hhi.info())

# Descriptif statistique
print("\nStatistiques descriptives :")
print(df_hhi.describe())

# Qualité des données
print("\nValeurs manquantes :")
print(df_hhi.isnull().sum())
print(f"Doublons détectés : {df_hhi.duplicated().sum()}")

# Vérification de la plage de l'indice
hhi_min = df_hhi['hhi_index'].min()
hhi_max = df_hhi['hhi_index'].max()
print(f"Plage de l'indice HHI : [{hhi_min:.4f} - {hhi_max:.4f}]")

# Identification des marchés les plus concentrés
print("\nLes 5 départements où la concurrence est la moins forte (HHI élevé) :")
display(df_hhi.nlargest(5, 'hhi_index'))

# Vérification des anomalies
if hhi_min < 0.25:
    print(" Attention : Des valeurs HHI sont inférieures au seuil théorique de 0.25.")
```

```python
infra_stats_chemin = os.path.join(parent, 'data\output\infra_stats.csv')
infra_stats_chemin_final = infra_stats_chemin.replace("\\", "/")

# Chargement
df_infra = pd.read_csv(infra_stats_chemin_final)

# en-tête
print("En-tête du DataFrame :")
print(df_infra.head())

# Infos structurelles
print(f"Dimensions : {df_infra.shape}")
print(df_infra.info())

# Descriptif statistique
print("\nStatistiques descriptives :")
print(df_infra.describe())

# Qualité des données
print("\nValeurs manquantes :")
print(df_infra.isnull().sum())
print(f"Doublons détectés : {df_infra.duplicated().sum()}")

# Vérification de cohérence (Total 4G ne peut pas être > Total Sites)
mask_error = df_infra['total_4g'] > df_infra['total_sites']
if mask_error.any():
    print("Erreur logique détectée : Des opérateurs ont plus de sites 4G que de sites totaux.")
else:
    print("Cohérence logique vérifiée (4G <= Total).")

# Calcul de la contribution relative
df_infra['poids_reseau_pct'] = (df_infra['total_sites'] / df_infra['total_sites'].sum()) * 100
print("\nPoids de chaque opérateur dans l'infrastructure nationale :")
display(df_infra[['nom_op', 'total_sites', 'poids_reseau_pct', 'taux_4g_pct']])
```

```python
matrice_concurrence_chemin = os.path.join(parent, 'data\output\matrice_concurrence.csv')
matrice_concurrence_chemin_final = matrice_concurrence_chemin.replace("\\", "/")

# Chargement
df_matrice = pd.read_csv(matrice_concurrence_chemin_final)

# en-tête
print("En-tête du DataFrame :")
print(df_matrice.head())

# Infos structurelles
print(f"Dimensions : {df_matrice.shape}")

# Qualité des données
print("\nValeurs manquantes :")
print(df_matrice.isnull().sum())
print(f"Doublons détectés : {df_matrice.duplicated().sum()}")

# Descriptif statistique
print("\nStatistiques descriptives :")
print(df_matrice.describe())

# Détection des doublons de nommage de régions
n_reg_raw = df_matrice['nom_reg'].nunique()
print(f"Nombre de noms de régions uniques avant nettoyage : {n_reg_raw}")

# Affichage des noms pour identifier les corrections nécessaires
print("\nListe des noms de régions bruts :")
print(df_matrice['nom_reg'].unique())

# Vérification de la complétude (présence de 0 ou de valeurs nulles)
print("\nNombre de valeurs à zéro par opérateur :")
print((df_matrice[['Bouygues Telecom', 'Free Mobile', 'Orange', 'SFR']] == 0).sum())

# Aperçu global
print("\nAperçu global du DataFrame :") 
display(df_matrice.head())
```
